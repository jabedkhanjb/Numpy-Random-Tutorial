# Normal Distribution

The Normal Distribution is one of the most important distributions.

It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.

It fits the probability distribution of many events, eg. IQ Scores, Heartbeat etc.

Use the <b>random.normal()</b> method to get a Normal Data Distribution.

It has three parameters:

<b>loc</b> - (Mean) where the peak of the bell exists.

<b>scale</b> - (Standard Deviation) how flat the graph distribution should be.

<b>size</b> - The shape of the returned array.

<i><b>Generate a random normal distribution of size nxn:</b></i>

```python
from numpy import random
x = random.normal(size = (4, 3))
print(x)
```

<i><b>Generate a random normal distribution of size n x n with mean at 1 and standard deviation of 2:</b></i>

```python
from numpy import random

x = random.normal(loc=1, scale=2, size=(2, 3))

print(x) 
```

# Visualization of Normal Distribution

```python
import matplotlib.pyplot as plt
from numpy import random
import seaborn as sb

sb.distplot(random.normal(size=1000), hist = False)

plt.show()
```
